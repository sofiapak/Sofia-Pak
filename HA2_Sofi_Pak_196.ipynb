**Task 3**

Demonstrate at least 2 ways to check if a given string is a palindrome. Your code should return True if this string is a palindrome and False otherwise. 

A palindrome is a word, number, phrase, string, or other sequence of characters which reads the same backward as forward. Examples: 'madam', 'kayak', 'aaabbbbaaa'.   

````python
candidate = 'aAaacbbcaaAa'

palindrome = ''.join(reversed(candidate))

if candidate == palindrome:
    print ('true')
else: 
    print ('false')
    
````

````python
candidate = 'aAaacbbcaaAa'
w=str(candidate)
p= len(w)
for i in range (p//2):
    if w[i] != w[-1-i]:
        print("not a palidrome")
        break
else: 
    print("this is a palindrome")
````

5.3. Create an empty list. For each number check the condition: if this number is divided by 3 (without the remainder), add a word 'Three' in a new list, if it is divided by 5, add a word 'Five', if it is divided by both 3 and 5 - add 'ThreeFive'. Else (if all these conditions are not satisfied), simply add this number to a new list.

The beginning of this sequence should be: [1, 2, 'Three', 4, 'Five', 'Three', 7, 8, 'Three', 'Five', 11, 'Three', 13, 14, 'ThreeFive', ...]

````python
>>> list(range (1,1001))
````

````python
empty_list=[]
n=1000
for i in range (n):
    a = i+1 
    if a % 3 == 0:
        if a % 5 == 0:
            empty_list.append('ThreeFive')
        else:
            empty_list.append('Three')
    elif a % 5 == 0:
        empty_list.append('Five')
    else:
        empty_list.append(a)
print(empty_list)
        
````
